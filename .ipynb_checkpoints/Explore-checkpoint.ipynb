## တစ်ရက်ချင်းစီအတွက် data ကိုကြည့်ရအောင်

```python
import datetime
from datetime import date
import re

daily_data=[]
for file in filenames:
    match = re.search(r'\d{2}-\d{2}-\d{4}', file)
    date = datetime.datetime.strptime(match.group(), '%d-%m-%Y').date()
    tmp_df = pd.read_csv(file).sum()
    tmp_df['Date'] = date
    daily_data.append(tmp_df)
cols = ['Date']+df_sum_by_region.columns.tolist()
```

```python
df_daily_data = pd.DataFrame(daily_data)[cols].sort_values(by=['Date']).fillna(0)
df_daily_data
```

## Cumulative Confirmed Plot (စုစုပေါင်း တွေ့ရှိအရေအတွက်)

```python
df_daily_data.plot(x ='Date', y='Confirmed', kind = 'line', figsize=(10,5))
```

# Number of confirmed for each day

```python
df_diff = df_daily_data[['Confirmed', 'Lab_Neg']].diff(axis=0).fillna(0)
df_diff["Total_tests"] = df_diff.loc[:,'Confirmed':'Lab_Neg'].sum(axis = 1)
df_diff.columns = map(lambda x: 'diff_'+x, df_diff)
df_diff
```

```python
df_daily_data=df_daily_data.join(df_diff, how='outer')
df_daily_data
```

## ၂၇ ရက်နေ့ကစပြီး တစ်ရက်ချင်းစီအတွက် တိုးလားတဲ့တွေ့ရှိအရေအတွက်

```python
#ax = plt.gca()
df_daily_data.plot(x ='Date', y='diff_Confirmed', kind = 'line', figsize=(10,5))
#df_daily_data.plot(x ='Date', y='diff_Lab_Neg', kind = 'line', figsize=(10,5), color='red',ax=ax)
```

## ၂၇ ရက်နေ့ကစပြီး တစ်ရက်ချင်းစီအတွက် test အရေအတွက်
Data ပဲမှားလို့ထင်ပါတယ်.. April ၁ရက်နေ့မှာ test နှစ်ခုပဲလုပ်တယ်ဆိုတာ မဖြစ်နိုင်ပါ

```python
df_daily_data.plot(x ='Date', y='diff_Total_tests', kind = 'line', figsize=(10,5))
```

## Medical supplies (ဆေးနှင့်ဆေးပစ္စည်းများ) data ကို လေ့လာရအောင်

မတ်လ ၂၇ရက်နေ့နောက်ပိုင်း ဖြန့်ဖြူးထားတဲ့ ဆေးနှင့်ဆေးပစ္စည်းအရေအတွက်ပါ.. မတ်လ ၂၇မတိုင်ခင်မှာရှိတဲ့အရေအတွက်က data ထဲမှာမပါပါ။

```python
df_hospital_names = pd.read_csv('Hospital Names in English and Burmese.csv')
df_hospital_names.columns=['hospital-name-en', 'hospital-name-mm']
```

```python
df_medical_supplies = pd.read_csv('Medical Supplies with Location.csv').fillna(0).iloc[:,:-3]
df_medical_supplies = pd.merge(df_medical_supplies, df_hospital_names, how='left', on=['hospital-name-mm'])
```

```python
df_medical_supplies
```

```python
total_num_ventilators = df_medical_supplies.sum()['Ventilators']
print("total_numbers_of_ventilators: ", total_num_ventilators)
```

```python
df_medical_supplies_by_hospital = df_medical_supplies.groupby(['Hospital-original','hospital-name-mm']).sum().reset_index()
df_medical_supplies_by_hospital
```

```python
df_medical_supplies_by_hospital.plot(x ='hospital-name-mm', y='PPE', kind = 'bar', figsize=(10,5))
```

```python
df_medical_supplies_by_hospital.plot(x ='hospital-name-en', y='Ventilators', kind = 'bar', figsize=(10,5))
```
